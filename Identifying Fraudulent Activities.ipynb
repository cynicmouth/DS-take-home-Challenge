# Challenge Description

Company XYZ is an e-commerce site that sells hand-made clothes. <br>
You have to build a model that predicts whether a user has a high probability of using the site to perform some illegal activity or not. This is a super common task for data scientists.
You only have information about the user first transaction on the site and based on that you have to make your classification ("fraud/no fraud"). <br>
These are the tasks you are asked to do: 
- For each user, determine her country based on the numeric IP address.
- Build a model to predict whether an activity is fraudulent or not. Explain how different assumptions about the cost of false positives vs false negatives would impact the model.
- Your boss is a bit worried about using a model she doesn't understand for something as important as fraud detection. How would you explain her how the model is making the predictions? Not from a mathematical perspective (she couldn't care less about that), but from a user perspective. What kinds of users are more likely to be classified as at risk? What are their characteristics?
- Let's say you now have this model which can be used live to predict in real time if an activity is fraudulent or not. From a product perspective, how would you use it? That is, what kind of different user experiences would you build based on the model output?

## Finding out `country` info given lower and upper bound of `ip_address`

## Feature Engineering Variables

### `signup_time` & `purchase_time`  time difference v.s. Fraud 

```python
fraud_data['diff_min'] = (fraud_data.purchase_time - fraud_data.signup_time).astype('timedelta64[m]')
fraud_data.head()
```

```python
sns.set(style = 'darkgrid')
fig, ax = plt.subplots(figsize = (30,10))
sns.distplot(fraud_data.diff_min, bins = 50 , kde = False)
ax.set_title('histogram of time difference in purchase')
ax.set_ylabel('Count Occurance')
```

```python
sns.set(style = 'darkgrid')
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (20,6))
sns.distplot(fraud_data[fraud_data['class']==0].diff_min, bins = 50, ax= ax[0], label = 'Class 0')
sns.distplot(fraud_data[fraud_data['class']==1].diff_min, bins = 50, ax= ax[0], label = 'Class 1')
ax[0].set_title('histogram of time difference in purchase')
ax[0].set_ylabel('Count Occurance')
ax[0].legend()
ax[0].set_title('KDE plot of time difference in purchase by Frauduent Class')
ax[0].set_ylabel('Kernel Density')


ax[1].plot('diff_min', 'class', '.-', data = fraud_data.groupby('diff_min')['class'].mean().reset_index())
```

Seems like the difference in time does have correlations to impact

```python
func_hour = lambda x: x.hour
func_day = lambda x: x.day
func_week = lambda x: x.week
func_month = lambda x: x.month

daytime_func = [func_hour, func_day, func_week, func_month]
daytime_func_name = ['hour', 'day', 'week', 'month']
daytime_func_dict = dict(zip(daytime_func_name, daytime_func))

for col in ['signup_time', 'purchase_time']:
    for name, func in daytime_func_dict.items():
        # setting up a group by count over col, func
        groupby_class0 = fraud_data[fraud_data['class']==0][col].apply(func)        
        groupby_class1 = fraud_data[fraud_data['class']==1][col].apply(func)
        groupby = fraud_data.groupby(fraud_data[col].apply(func))['class'].mean().reset_index()
        x= np.array(groupby[col])
        y= np.array(groupby['class'])
        fit = np.polyfit(x,y,5)
        fit_func = np.poly1d(fit)
        
        
        # plotting results using groupby
        fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (20,6))
        sns.distplot(groupby_class0, ax = ax[0], label = 'Class 0')
        sns.distplot(groupby_class1, ax = ax[0], label = 'Class 1')
        ax[0].legend()
        ax[0].set_title('Fraud Class Distribution of {} by {}'.format(col, name))
        
        ax[1].plot(groupby[col], groupby['class'], '.-', label = 'mean fraud rate')
        ax[1].plot(x, fit_func(x), '--', label = 'fraud rate trend')
        ax[1].legend()
        ax[1].set_title('Fraud Class Rate of {} by {}'.format(col, name))
```

Also if we break it down by `signup_time` and `purchase_time`, we can see that:
- thers is not strong correlation in hour
- however there seems to have a strong correlation on day, week and month, as beginning of those ten to have a higher fraud rate, so i think we can put some feature engineering there

### `device_id` v.s. Fraud

Let's also create a variable to indicate if the device was used to make multiple purchases, i.e. shared by different users and see if this can be a factor contributing to fraud

```python
fraud_data = pd.merge(fraud_data, 
                      fraud_data.groupby('device_id')['user_id'].count().reset_index(), 
                      how = 'left', on = 'device_id', suffixes = ('','_y')
                     ).rename(columns = {'user_id_y':'transactions_on_device'})

fraud_data.head()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 6))
sns.distplot(fraud_data[fraud_data['class'] == 0].transactions_on_device, ax = ax[0], label = 'Class 0')
sns.distplot(fraud_data[fraud_data['class'] == 1].transactions_on_device, ax = ax[0], label = 'Class 1')
ax[0].legend()
ax[1].set_title('distribution of transactions_on_device')

groupby = fraud_data[['transactions_on_device','class']].groupby('transactions_on_device').mean().reset_index()
x= np.array(groupby['transactions_on_device'])
y= np.array(groupby['class'])
fit = np.polyfit(x,y,5)
fit_func = np.poly1d(fit)

ax[1].plot(x, y, '.-', label = 'mean fraud rate')
ax[1].plot(x, fit_func(x), '--', label = 'fraud rate trend')
ax[1].set_title('fraud rate by transactions_on_device')
```

Clearly, this is also a good feature to have as more transactions placed on the same device tends to be fraudulent, which make sense

## Building Classifier with H2O

```python
import h2o
from h2o.frame import H2OFrame
from h2o.estimators.random_forest import H2ORandomForestEstimator

h2o.init()
h2o.remove_all()
```

### Loading Data into H2O Environment

```python
h2o_df = H2OFrame(fraud_data_ml)
h2o_df[['source','browser', 'sex', 'country', 'class']] =  h2o_df[['source','browser', 
                                                                   'sex', 'country', 'class']].asfactor()
h2o_df.summary()
```

### Train Test Split

```python
strat_split = h2o_df['class'].stratified_split(test_frac = 0.25, seed = 123)

train = h2o_df[strat_split == 'train']
test = h2o_df[strat_split == 'test']

feature = [ col for col in list(fraud_data_ml.columns) if col != 'class']
target = 'class'

model = H2ORandomForestEstimator(balance_classes = True, ntrees = 500, max_depth = 50, seed = 123, 
                                 score_each_iteration = True, nfolds = 5)
model.train(x = feature, y=target,  training_frame = train)
```
